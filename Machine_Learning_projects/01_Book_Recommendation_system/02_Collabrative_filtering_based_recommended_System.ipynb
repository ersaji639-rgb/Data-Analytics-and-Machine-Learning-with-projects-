```python
import numpy as np 
import pandas as pd
```

### Collaborating based

* Users -- we choose such a way that they have gave rating for ore than 200 books, 
* Books -- we consider such that, it have min 50 ratings

```python
books = pd.read_csv("Books.csv")
users = pd.read_csv("Users.csv")
ratings = pd.read_csv("Ratings.csv")
```

```python
# Mergeing two dataframes using ISBM col.
ratings_with_name = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200 
userID_rated_gt200_books =  x[x].index
userID_rated_gt200_books
```

```python
# filtered Based on users
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(userID_rated_gt200_books)]
```

```python
# filtered based on ratings 
y=filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books =y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
# Making a pivot table
pt = final_ratings.pivot_table(index = "Book-Title", columns = "User-ID", values = "Book-Rating")
pt.shape
```

```python
pt.fillna(0, inplace = True) 
```

```python
pt
```

```python
# Calculate the distance: 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt) # Eucledian distance 
```

##### `sorted(..., key=lambda x: x[1])`

- Here, `x` represents each tuple of the form **(index, similarity_value)**.  
- `x[1]` corresponds to the **similarity value** inside the tuple.  
- Therefore, the sorting is done **based on the similarity score**, not the index.


```python
# Sort based on the distance 

sorted(list(enumerate(similarity_scores[0])) , key = lambda x:x[1], reverse = True)[1:6]  #Here first one is this bo itseld 
```

```python
def recommend(book_name):
    # Index fetch 
    index =  np.where(pt.index == book_name )[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])) , key = lambda x:x[1], reverse = True)[1:6] 
    for i in similar_items:
        print(pt.index[i[0]])
   
```

```python
# Check 2:
recommend("1984")
```
