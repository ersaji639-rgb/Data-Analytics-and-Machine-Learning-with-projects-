```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
print(df_tracks.head(5))
```

```python
# Null values check in each column. 
pd.isnull(df_tracks).sum()
```

```python
print(df_tracks.info())
```

#### # Top 10 least popular songs: 

```python
# Top 10 least popular songs: 
soarted_df = df_tracks.sort_values('popularity',ascending= True ).head(10)
print(soarted_df)
```

```python
df_tracks.describe().transpose()
```

#### top 10 songs with popularity greater than 90. 

```python
most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending= False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date", inplace = True)
```

```python
df_tracks.index = pd.to_datetime(df_tracks.index) # Change from obj to data=e time format
```

```python
df_tracks.head()
```

```python
# to filter speecific row from the data  set. say(18 row)  
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration_ms"].head()
```

```python
# Convert the duration from millisec to sec:

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks["duration"].head()
```

####  correlation maps

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis = 1).corr(method= "pearson")
```

```python
plt.figure(figsize= (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidth = 1, linecolor = "Black")
heatmap.set_title("Correlationheat map between var")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
len(df_tracks)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
len(sample_df)
```

```python
# We are trying to make a regression plot betweeen loudness and energy, because we saw a high positive corelation between those from the above. 
plt.figure(figsize=(10,6))
ax = sns.regplot(
    data=sample_df, 
    x="energy", 
    y="loudness",
    scatter_kws={"s": 50, "alpha": 0.6},  # marker size & transparency
    line_kws={"color": "red", "lw": 2},   # regression line style
    color="c"
)

ax.set_title("Loudness vs Energy Correlation", fontsize=14, weight="bold")
ax.set_xlabel("Energy", fontsize=12)
ax.set_ylabel("Loudness", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
sample_df.columns
```

```python
# Regression plot between two features:
plt.figure(figsize=(10,6))
ax = sns.regplot(
    data=sample_df, 
    x="acousticness", 
    y="popularity",
    scatter_kws={"s": 50, "alpha": 0.6},  # marker size & transparency
    line_kws={"color": "red", "lw": 2},   # regression line style
    color="c"
)

ax.set_title("Popularity vs Acousticness", fontsize=14, weight="bold")
ax.set_xlabel("Popularity", fontsize=12)
ax.set_ylabel("acousticnes", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values("release_date")
```

```python
df_tracks["dates"].head()
```

```python
df_tracks.dates.head()
```

```python
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year

# Or 
#df_tracks["dates"] = pd.to_datetime(df_tracks["dates"])
#df_tracks["year"] = df_tracks["dates"].dt.year
```

```python
years.head()
```

#### Distribution plot for total number of songs availabe in the spotify since 1992. 

```python
sns.displot(
    years, 
    discrete=True, 
    aspect=2, 
    height=5, 
    kind="hist"
).set(title="No of Songs per Year")
```

#### Duration of songs across year



```python
total_dr = df_tracks["duration"]
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=years, y=total_dr, ax=ax, errwidth=0)  # errwidth=0 hides error bars

ax.set_title("Year vs Duration")
plt.xticks(rotation=90)

plt.show()
```

```python
# In line plt 
total_dr  = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x= years, y = total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```

### Second Data : Includes movie genre:

```python
df_genre = pd.read_csv("SpotifyFeatures.csv") 
df_genre.head()
```

```python
df_genre.genre.unique()
```

```python
df_genre["genre"].value_counts()
```

```python
df_genre.genre.nunique()
```

```python
plt.title("Duration of the songs in differen genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y = 'genre', x = "duration_ms", data = df_genre)
plt.xlabel("Duration in milli Seconds")
plt.ylabel("genres")
```

###  Top 5 Generes by popularity 

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))

famous = df_genre.sort_values("popularity", ascending=False).head(10)  # top 5

ax = sns.barplot(y="genre", x="popularity", data=famous)
ax.set(title="Top 5 Genres by Popularity")

plt.show()
```
