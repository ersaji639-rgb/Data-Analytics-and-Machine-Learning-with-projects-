### Data Cleaning

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({"LF":"Low Fat", 
                                                         "low fat":"Low Fat",
                                                         "reg":"Regular"})
```

```python
print(df['Item Fat Content'].unique())
```

## Business Requirements:

### KPI requirements:

```python
# Total Sales:
total_sales = df['Sales'].sum()

# Average Sales 

avg_sales = df['Sales'].mean()

#Number of items sold 
no_of_item_sold = df['Sales'].count()

# Avg Ratings         

avg_ratings = df['Rating'].mean()


# Display
print(f"Total Sales: ${total_sales:,.1f}" ) # comma for 1000 separator. 
print(f"Average Sales: ${avg_sales:,.1f}" )
print(f"Number of items sold: {no_of_item_sold:,.1f}" )
print(f"Average Ratings: {avg_ratings:,.1f}" )
```

### Charts: 

### Total Sales by Fat content:

```python
sales_by_fat = df.groupby("Item Fat Content")['Sales'].sum()
print(sales_by_fat)
fig = plt.figure(figsize=(3,3))
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%.1f%%',
       startangle = 90) # %% is a place holder of numbers

plt.title("Sales by fat content") 

plt.axis("equal")

plt.show()
```

### Total Sales by item type

```python
df.columns
```

```python
df['Item Type'].unique()
```

```python
sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)
print(sales_by_item_type)
```

```python
plt.figure(figsize = (10,6))
bars = plt.bar(sales_by_item_type.index, sales_by_item_type.values)

plt.xticks(rotation = -90)

plt.xlabel('Item Type')

plt.ylabel('Total Sales')
plt.title("Total Sales by item type")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() /2, bar.get_height(),
            f'{bar.get_height():,.0f}', ha = 'center', va='bottom', fontsize = 8)

plt.tight_layout()
plt.show()







```

### Fat content by outlet for sales

```python
df.columns
```

```python
df[ 'Outlet Type'].unique()
```

```python
grouped = df.groupby(['Outlet Location Type','Item Fat Content' ])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]
```

```python
ax = grouped.plot(kind='bar', figsize=(8,5), title='Outlet Tier by Item Fat Content')

plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(
        str(int(p.get_height())),                 # convert height to int
        (p.get_x() + p.get_width() / 2, p.get_height()),  # x,y position
        ha='center', va='bottom', fontsize=9, color='black'
    )

plt.tight_layout()
plt.show()
```

### Total Sales by outlet establishment:


```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize = (8,5))
plt.plot(sales_by_year.index, sales_by_year.values, marker = 'o', linestyle = '-')

plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")

plt.title("Total Sales by outlet establishment")

for x, y in zip(sales_by_year.index, sales_by_year.values ):
    plt.text(x,y,f'{y:,.0f}', ha = 'center', va = 'bottom', fontsize = 8)

plt.tight_layout()
plt.show()
```

### Sales by outlet size:


```python
sales_by_outlet_size = df.groupby('Outlet Size')['Sales'].sum()
```

```python
sales_by_outlet_size
```

```python
plt.figure(figsize=(3,3)) 
plt.pie(sales_by_outlet_size, labels= sales_by_outlet_size.index, autopct='%1.1f%%', startangle= 90)
plt.title("Sales based on outlet Size")
plt.tight_layout()
plt.show()
```

### Sales by outlet location:

```python
sales_by_outlet_location = df.groupby("Outlet Location Type")['Sales'].sum().reset_index()
```

```python
sales_by_outlet_location = sales_by_outlet_location.sort_values('Sales', ascending= False)
sales_by_outlet_location
```

```python
plt.figure(figsize=(8,3))
ax = sns.barplot(
    x='Sales', 
    y='Outlet Location Type', 
    data=sales_by_outlet_location
)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

# Add labels on each bar
for p in ax.patches:
    width = p.get_width()   # bar length (since x-axis is Sales)
    plt.text(
        width + (0.01 * sales_by_outlet_location['Sales'].max()),  # a little to the right of the bar
        p.get_y() + p.get_height() / 2,   # vertically centered
        f'{int(width):,}',                # format with commas
        ha='left', va='center'
    )

plt.tight_layout()
plt.show()
```
