### Drop unnamed column

```python
# drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

### Gender distribution

```python
# To count the gender in the school
plt.figure(figsize = (5,5))
ax = sns.countplot(data  = df, x =  'Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

#### From the above chart we have analysed that number of females is more than the number of males in the data from the school

### Impact of parents education on student score

```python
print(df.head())
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore" : 'mean', "ReadingScore" : "mean","WritingScore" : "mean"}) # Using groupby 
print(gb)
```

```python
plt.figure(figsize = (4,4))
plt.title("Gender Distribution")
sns.heatmap(gb, annot = True)
plt.show()
```

#### From the above chart it is understood that education of the parents has a good impact of students score

### How marital status of parents affect the mark of student 

```python
df.columns
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" : 'mean', "ReadingScore" : "mean","WritingScore" : "mean"}) # Using groupby 
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between the parents marital status and student score")
plt.show()
```

#### From above plot, Parents marital staus has less impact on student score.

### Impact on marks for the Student who play sports 

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore" : 'mean', "ReadingScore" : "mean","WritingScore" : "mean"}) # Using groupby 
print(gb2)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb2, annot = True)
plt.title("Relationship between the sports and student score")
plt.show()
```

#### Sports have good impact on marks
