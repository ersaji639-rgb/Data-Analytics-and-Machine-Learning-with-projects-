```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('sales', axis = 1)
y = df['sales']
```

```python
X
```

### Polynomial conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p_conv = PolynomialFeatures(degree= 3, include_bias= False)
```

```python
p_feat = p_conv.fit_transform(X)
```

```python
p_feat.shape # Converted into 19 features
```

### Train test split 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(p_feat, y, test_size = 0.25, random_state= 101)
```

### Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler() # We are performing standadization: which gives a rescaled data which has a mean of 0 and std of 1. 
```

```python
sc.fit(X_train) # Caution: only for train data 
```

```python
sc_X_train = sc.transform(X_train)
sc_X_test = sc.transform(X_test)
```
