### 📊 Visualizing MSE, Variance, and R² with Python

We will:
- Generate simple synthetic data
- Fit a linear regression model
- Compute and compare:
  - **MSE**: How far predictions are from true values
  - **Variance**: How spread out the actual target is
  - **R²**: How much variance is explained by the model


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create simple synthetic data
np.random.seed(0)
# Generate 50 evenly spaced numbers between 0 and 10, and reshape into a column vector (50 rows, 1 column)
X = np.linspace(0, 10, 50).reshape(-1, 1)
# Create the corresponding y-values using a linear function (slope=3, intercept=4) 
# and add random Gaussian noise to simulate real-world data
y_true = 3 * X.flatten() + 4 + np.random.normal(0, 4, size=50)

# Fit linear regression model
model = LinearRegression()
model.fit(X, y_true)
y_pred = model.predict(X)

# Metrics
mse = mean_squared_error(y_true, y_pred)
var_y = np.var(y_true)
r2 = r2_score(y_true, y_pred)

print(f"MSE       : {mse:.2f}")
print(f"Var(y)    : {var_y:.2f}")
print(f"R²        : {r2:.4f}")
print(f"MSE / Var : {mse / var_y:.4f}")
print(f"1 - (MSE / Var) = R²: {1 - (mse / var_y):.4f}")

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y_true, label="True Data", color="blue", alpha=0.6)
plt.plot(X, y_pred, label="Model Prediction", color="red", linewidth=2)
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

###  So Var(y) says how my actual y(label) is distributed from its mean)
 In other words:

   * If all your y values are very close to their mean, then variance is low.

 *If your y values are spread out (highly scattered), then variance is high.

### 📊 What Does Variance of \( y \) Mean?

---

### ✅ Understanding Variance of \( y \)

> **Variance of \( y \)** tells you how much the actual target values (labels) deviate from their mean.

- If values are **close to the mean** → **low variance**
- If values are **spread out** → **high variance**

---

### 📐 Formula

\[
\text{Var}(y) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \bar{y})^2
\]

Where:
- \( y_i \): Actual label values
- \( \bar{y} \): Mean of the actual labels

This measures the **average squared distance** of each label from the mean.

---

### 🧠 Why Variance Matters

- Variance gives a sense of how **spread out** or **noisy** your dataset is.
- It helps you understand **how hard** the prediction task is:
  - **Low variance**: Simple model might work well.
  - **High variance**: Model must capture more complexity.

---

### 🔗 Connection to R²

R² compares your model’s prediction error (MSE) to how much variance is in the data:

\[
R^2 = 1 - \frac{\text{MSE}}{\text{Var}(y)}
\]

- If **MSE is much smaller than Var(y)** → \( R^2 \approx 1 \) → ✅ Good model
- If **MSE ≈ Var(y)** → \( R^2 \approx 0 \) → ❌ Model no better than predicting the mean

---

### 🎯 Summary

| Concept        | Meaning |
|----------------|---------|
| **Variance of \( y \)** | How much actual values spread out from the mean |
| **Low variance**        | Data is tightly clustered around the mean |
| **High variance**       | Data is widely spread |
| **Used in**             | R² metric to evaluate model quality |
