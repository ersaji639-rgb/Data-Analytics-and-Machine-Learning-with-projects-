###  Or 
* Building a vocabulary (creating a bag of words )

```python
with open('One.txt') as mytext:
    words_one = mytext.read().lower().split()
```

```python
words_one
```

```python
len(words_one)
```

```python
unique_words_one = set(words_one)
```

```python
unique_words_one
```

```python
with open('Two.txt') as mytext:
    words_two = mytext.read().lower().split()
```

```python
words_two
```

```python
len(words_two)
```

```python
unique_words_two = set(words_two)
```

```python
unique_words_two
```

### Get all unique word :

```python
all_uni_words = set()
all_uni_words.update(unique_words_one)
```

```python
all_uni_words.update(unique_words_two)
```

```python
all_uni_words
```

### Assigning numbers for each of this unique words:

```python
full_vocabulary = dict()
i = 0 
for w in all_uni_words:
    full_vocabulary[w] = i
    i = i + 1
```

```python
full_vocabulary
```

###  Bag of words to Frequency Counts:

```python
one_freq= [0]*len(full_vocabulary)
two_freq= [0]*len(full_vocabulary)
all_words = ['']*len(full_vocabulary) # bunch of wmpty string 
```

```python
for w in full_vocabulary:
    w_ind = full_vocabulary[w] # calling by key 
    all_words[w_ind] = w
```

```python
all_words
```

```python
# add in the count per word in the doc 

with open('One.txt') as f:
    one_text = f.read().lower().split()  # list 


for w in one_text:
    w_ind = full_vocabulary[w]
    one_freq[w_ind] += 1
    
```

```python
one_freq
```

```python
# add in the count per word in the doc 

with open('Two.txt') as f:
    two_text = f.read().lower().split()  # list 


for w in two_text:
    w_ind = full_vocabulary[w]
    two_freq[w_ind] += 1
    
```

```python
two_freq
```

###  Feature Extraction with sklearn.


```python
text = ["This is a line",
        "This is another line",
        "Completely different line"
       ]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer, CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
cv.fit_transform(text)
```

```python
sp_mat = cv.fit_transform(text)
```

```python
sp_mat.todense()
```

```python
cv.vocabulary_  # note here a is not there, since it is common word
```

```python
cv =  CountVectorizer(stop_words= 'english') # to ignore more stop words. 
```

```python
cv.fit_transform(text).todense()
```

```python
cv.vocabulary_
```

### Tfidf:

```python
# Tfidf transformer 
tfidf_transformer = TfidfTransformer()
```

```python
cv = CountVectorizer()
```

```python
counts = cv.fit_transform(text)
```

```python
counts
```

```python
cv.vocabulary_
```

```python
tfidf = tfidf_transformer.fit_transform(counts)
```

```python
tfidf.todense() # This gives the weights of the words. 
```
