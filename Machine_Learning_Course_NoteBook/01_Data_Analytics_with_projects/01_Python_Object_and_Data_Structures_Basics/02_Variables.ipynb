### Mutable vs Immutable 

```python
# Immutable types (cannot be changed)
x = 10   # int
y = "Hello"  # str
z = (1, 2, 3)  # tuple

# Mutable types (can be changed)
my_list = [1, 2, 3]  # list
my_dict = {"name": "Saji", "age": 25}  # dictionary

# Changing mutable types
my_list.append(4)
my_dict["age"] = 26

print(my_list)  # Output: [1, 2, 3, 4]
print(my_dict)  # Output: {'name': 'Saji', 'age': 26}
```

### Variable Aliasing 

```python
a = [1, 2, 3]
b = a  # b points to the same list as a

b.append(4)
print(a)  # Output: [1, 2, 3, 4] (both a and b changed!)
```

Memory Explanation

    a = [1, 2, 3]

        A list [1, 2, 3] is created in memory.

        The variable a stores a reference to that list.

    b = a

        b does NOT create a new list.

        Instead, b stores the same reference as a.

        Now, both a and b point to the same list in memory.

    b.append(4)

        Since lists are mutable, modifying b also modifies a because they refer to the same object.

### To over come above (Create an indep copy)

```python
a = [1, 2, 3]
b = a.copy()  # Creates a new list
#or 
b = list(a)

b.append(4)

print(a)  # Output: [1, 2, 3]  (Unchanged)
print(b)  # Output: [1, 2, 3, 4] (Only b changed)
```

### is vs == (compwring Variables) 

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a #Note : it actuall stores the memory, it doesnt create new list. 

print(a == b)  # True (Values are the same)
print(a is b)  # False (Different memory locations)
print(a is c)  # True (Same memory location)
```

```python
print(id(a), id(b))  # Different memory addresses
```

```python
x = 100
y = 100
print(x is y)  # True (same memory location)

x = 1000
y = 1000
print(x is y)  # False (Python creates separate objects)
```

✅ Python optimizes small integers (between -5 and 256) by storing them in a shared memory pool.
✅ This means x and y point to the same memory address for values in this range

```python
x = 100
y = 100
print(id(x), id(y))  # Same memory address
```

```python
x = 1000
y = 1000
print(x is y)  # False
```

Why is this False?

👉 Python does not cache integers above 256, so each assignment creates a new object in memory.
Small integers (-5 to 256) are cached, so variables pointing to the same value will have the same memory address.

Larger numbers (outside the range) are stored as separate objects in memory.

The is operator checks memory location, while == checks value.

### == Vs   is 

```python
x = 1000
y = 1000

print(x == y)  # True (same value)
print(x is y)  # False (different memory locations)
```

this optimization only applies to certain immutable objects, specifically:
✅ Cached (Shared Memory)

    Small integers (-5 to 256)

    Interned strings (short strings or identifiers like "hello" but not "a long string")

    Some boolean values (True, False)

    Some small tuples (depending on their contents)

❌ Not Cached (Separate Memory)

    Large integers (> 256 or < -5)

    Floats (1.5, 3.14 → always stored separately)

    Lists ([1, 2, 3] → always stored separately)

    Dictionaries ({"a": 1, "b": 2} → always stored separately)

    Sets & custom objects
