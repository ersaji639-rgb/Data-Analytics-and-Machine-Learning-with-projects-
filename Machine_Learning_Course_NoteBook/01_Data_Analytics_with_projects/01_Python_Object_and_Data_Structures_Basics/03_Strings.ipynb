###  String Formatting 

```python
' Insert another string with curly brackets: {}'.format('The inserted string')
```

```python
'I have {}, {}, and {}'.format('Dog','Cat','Tiger')
```

```python
'I have {y}, {z}, and {x}'.format(x= 'Dog',y = 'Cat',z = 'Tiger')
```

```python
name = 'Saji'
len_name = len(name)

'My name is {n} and number of char in my name is {l}'.format(n = name, l = len_name)
```

```python
#'My name is {n} and number of char in my name is {l}'.format(n = 'Saji', l = len(n)) # it threrror, hence we can usef-literals. 
```
