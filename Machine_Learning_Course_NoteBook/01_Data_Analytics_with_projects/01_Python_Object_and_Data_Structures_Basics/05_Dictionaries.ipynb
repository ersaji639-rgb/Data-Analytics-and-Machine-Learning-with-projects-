### Nested Dictionary

```python
d1 = {'k1': 123,
      'k2':[12,22,'Hello',['world']],
      'p1':'Shreyas',
      'courses': {'c1':'Deeplearning',
                  'c2':['Machine Learning']}
     }
```

```python
# To grab C2 
d1['courses']['c2'][0]
```
