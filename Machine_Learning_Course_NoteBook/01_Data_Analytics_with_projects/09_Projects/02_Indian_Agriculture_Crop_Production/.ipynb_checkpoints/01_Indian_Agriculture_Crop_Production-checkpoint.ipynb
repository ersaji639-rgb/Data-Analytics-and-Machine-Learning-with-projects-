###  Dealing with different production Units: 

```python
agri_df['Production Units'].value_counts()
```

```python
# Assume one cocunut is 2kg 
agri_df['Production Units'] == 'Nuts'
```

```python
agri_df[agri_df['Production Units'] == 'Nuts']
```

```python
agri_df[agri_df['Production Units'] == 'Nuts']['Crop']
```

```python
agri_df[agri_df['Production Units'] == 'Nuts']['Crop'].value_counts()  
```

So nuts units is used only for cocunut here.  (  from above) 

```python
# Convert the nuts to kgs 
def unit_conversion(df):
    if df['Production Units'] == 'Nuts':
        return df['Production']/50
    elif df['Production Units'] == 'Tonnes':
        return df['Production']
    else:
        return df['Production']/4.166
        
```

```python
agri_df['new production'] = np.round(agri_df.apply(unit_conversion,axis = 1),2)
```

### Highest producer of rice since 1997

```python
agri_df[agri_df['Crop'] == 'Rice']
```

```python
agri_rice = agri_df[agri_df['Crop'] == 'Rice']
agri_rice['State'].unique()
```

```python
unique_rice_growing_states = agri_rice['State'].unique()
```

```python
total_rice_production = []
for state in unique_rice_growing_states:
    #print(agri_rice.loc[agri_rice['State'] == state, 'new production'].sum())
    total_rice_production.append(agri_rice.loc[agri_rice['State'] == state, 'new production'].sum())
```

```python
total_rice_production
```

```python
rice_crop_production_statewise = pd.DataFrame({"State":unique_rice_growing_states,
"Total Rice Production" : total_rice_production})
rice_crop_production_statewise
```

```python
# I want highest one 
rice_crop_production_statewise.sort_values('Total Rice Production', ascending = False)
```

```python
print(f"Top 10 rice producing states in (Tonnes) are: \n {(rice_crop_production_statewise.sort_values('Total Rice Production', ascending = False)).head(10)}")
```

### Statewise total production of crops:

```python
unique_states = agri_df['State'].unique()
total_production_list = []
for state in unique_states:
    total_production_list.append(agri_df.loc[agri_df['State']==state, 'new production'].sum())

crop_production_statewise = pd.DataFrame({"State":unique_states,
                                         "Total crop production": total_production_list})
crop_production_statewise
```

### Best year for Agriculture 

```python
agri_df['Year'].unique()
```

```python
# Convert this array to list 
list(agri_df['Year'].unique())
```

```python
unique_year_list = list(agri_df['Year'].unique())
```

```python
year_wise_production = []
for year in unique_year_list:
    year_wise_production.append(agri_df.loc[agri_df['Year'] == year, 'new production'].sum())

crop_production_yearwise = crop_production_yearwise = pd.DataFrame({"year":unique_year_list,
"Total Crop Production":year_wise_production})

# Sort the values in decending order
crop_production_yearwise.sort_values('Total Crop Production', ascending = False).head(5)
```

```python
# Worse 
crop_production_yearwise.sort_values('Total Crop Production', ascending = True).head(5)
```
