```python
df.groupby('year').mean()
```

```python
df.groupby('year', as_index = False).mean()
```

```python
yearwise = df.groupby('year', as_index = False) # groupby yearwise
yearwise.mean()
```

```python
plt.figure(figsize = (14,6))
sns.barplot(x = 'year', y = 'Total', data = yearwise.mean())
```

```python
yearwise.mean()
```

```python
yearwise.mean().plot(x = 'year', y = ['POONDI',	'CHOLAVARAM'	,'REDHILLS'	,'CHEMBARAMBAKKAM'], figsize = (20,8))    # Frm pandsas,
```

```python
# Variarion with month 
month = df.groupby('month', as_index = False )
month.mean()
```

```python
plt.figure(figsize =(14,6))
sns.barplot(x = 'month', y = 'Total', data = month.mean())
plt.show()
```

```python
month.mean().plot(x = 'month', y = ['POONDI',	'CHOLAVARAM'	,'REDHILLS'	,'CHEMBARAMBAKKAM'], figsize = (20,8))    # Frm pandsas,
```
