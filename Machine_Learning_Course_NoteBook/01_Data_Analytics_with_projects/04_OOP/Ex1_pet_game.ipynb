🎮 Mini Project: “Adopt a Pet” Game
🐶 🐱 🐦 What You'll Build:

    The player chooses a pet (Dog, Cat, or Bird)

    The pet has a name and age

    You can interact with your pet: make it speak or celebrate its birthday

    🧠 Concepts You Practiced:

    Classes & objects

    Inheritance

    User input with input()

    Conditional logic (if, elif, else)

    Loops (while True)

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"\n{name} the animal is ready for adoption!")

    def speak(self):
        print(f"{self.name} makes a sound.")

    def birthday(self):
        self.age += 1
        print(f"🎂 {self.name} just turned {self.age}!")

class Dog(Animal):
    def speak(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} says Meow!")

class Bird(Animal):
    def speak(self):
        print(f"{self.name} says Chirp!")

# ---- Game starts here ----
print("Welcome to the Pet Adoption Center!")
pet_type = input("Choose a pet to adopt (dog / cat / bird): ").lower()
name = input("What would you like to name your pet? ")
age = int(input("How old is your pet? "))

# Create the pet based on user choice
if pet_type == "dog":
    pet = Dog(name, age)
elif pet_type == "cat":
    pet = Cat(name, age)
elif pet_type == "bird":
    pet = Bird(name, age)
else:
    print("Sorry, we don't have that pet.")
    exit()

# Game loop
while True:
    print("\nWhat would you like to do?")
    print("1. Make your pet speak")
    print("2. Celebrate a birthday")
    print("3. Exit")
    choice = input("Enter your choice (1/2/3): ")

    if choice == "1":
        pet.speak()
    elif choice == "2":
        pet.birthday()
    elif choice == "3":
        print(f"Goodbye! {pet.name} will miss you 🐾")
        break
    else:
        print("Please enter a valid option.")
```
