####  Line accepts coordinate as a pair of tuples and return the slope and distance of the line 

```python
class Line:
    # Attribute 
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    # Method 1 
    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2 
        return ((x2 -x1)**2 + (y2-y1)**2)**0.5
    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2 
        return (y2 -y1)/(x2-x1)
        
        
```

```python

l1 = Line(coor1 = (3,2), coor2 = (9,5)) # or Line((3,2),(9,5))
```

### Inheritance
It helps to create new classes using the classes which are already defined. 
* Forms classes based on already existing classes
* The newly formed classes are called child classes / Derived classes.
* The already exixting class are called base class/ parent class.
* The derive class (descendants) over ride or extend the functionality of the base class (ancestors). 

```python
class  Animal:
    def __init__ (self):
        print ("Animal created")
    def whoAmI(self):
        print("Iam a Animal")
    def eat(self):
        print("Iam Eating")
```

```python
class  Animal:
    def __init__(self):
        print ("Animal created")
    def whoAmI(self):
        print("Iam a Animal")
    def eat(self):
        print("Iam Eating")
# Now Iam creating one more class dog which is going to inherit all the properties the Animal class has. 
class Dog(Animal): 
    def __init__(self):
        Animal.__init__(self) # it is is inheriting all the attributesf animal class. 
        print("Dog ")       
    def whoAmI(self): # Note: the method will be similar as animal class. ( This will over write this method for dog class. 
        print("Iam a Dog") 
    def bark(self):
        print("Woof")

   
        
```

```python
class Animal:
    def __init__(self):
        print("Animal created")

    def whoAmI(self):
        print("I am an Animal")

    def eat(self):
        print("I am Eating")

class Dog(Animal): 
    def __init__(self):
        super().__init__()  # Preferred way in modern Python to call the base constructor
        print("Dog created")

    def whoAmI(self):  # Overrides Animal's whoAmI method
        print("I am a Dog")

    def bark(self):
        print("Woof")

# Example usage
d = Dog()
d.whoAmI()
d.eat()
d.bark()
```

### Polymorphisim
*it is the way in which different object classes share the same method name. 

```python
class Dog:
    def __init__(self,name):
        self.name = name

    def sound(self):
        return f"{self.name} says woof"
class Cat:
    def __init__(self,name):
        self.name = name

    def sound(self):
        return f"{self.name} says Meow"
```

```python
dog1 = Dog(name = "Niko")
```

```python
dog1.sound()
```

```python
cat1 = Cat(name = "Tom")
```

```python
cat1.sound()
```

```python
# Polymorphism using for loop

for pet in  [dog1,cat1]:
    print(pet.sound())
```

```python
# polymorphism using functions 

def pet_speak(pet):
    print(pet.sound())
```

```python
pet_speak(cat1)
```

```python
pet_speak(dog1)
```

### Special methods

```python
class Book:
    def __init__(self,title,author,pages):
        self.title = title
        self.author = author 
        self.pages  = pages 

    def __str__(self): # speial method :string
        return f"Title:{self.title},\n Author : {self.author},\n Pages : {self.pages}" 


    def __len__(self): # to get number 
        return self.pages

    def __del__(self):
        print("A book is burned")
```

```python
# Example 
class BankAccount:
    def __init__(self, account_holder, balance):
        self.owner = account_holder
        self.balance = balance
    def __str__(self):
        return f"Account owner: {self.owner}\n Account Balance : ${self.balance}"
    def deposit(self,dep_amt):
        self.balance += dep_amt
        print(f"${dep_amt} deposited") 

    def withdraw(self, wd_amt):
        if self.balance >= wd_amt: 
            self.balance -= wd_amt
            print("Withdrawal accepted")
        else:
            print("insufficient funds")
```

```python
c1 = BankAccount(account_holder= "Saji", balance= 100)
```

```python
print(c1)
```

```python
c1.deposit(dep_amt=2000)
```

```python
print(c1)
```
