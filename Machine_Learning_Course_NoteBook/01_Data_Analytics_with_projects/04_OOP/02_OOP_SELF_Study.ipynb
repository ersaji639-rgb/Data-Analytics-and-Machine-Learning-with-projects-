🧱 Basic Concepts of OOP

Think of OOP as modeling things from the real world into code using:

    Class – A blueprint (like a recipe)

    Object – A specific instance (like the actual dish made from the recipe)

    Attributes – What an object has (e.g., name, age)

    Methods – What an object can do (e.g., speak, walk)

#### Simple class

```python
class Person:
    def __init__(self, name, age):  # constructor
        self.name = name            # attribute
        self.age = age

    def greet(self):                # method
        print(f"Hi, I am {self.name} and I am {self.age} years old.")

# Create an object
p1 = Person("Saji", 25)

# Call method
p1.greet()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal '{self.name}' created")

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
        print(f"Dog '{self.name}' created")

    def speak(self):
        print(f"{self.name} says Woof!")

# Test it
a = Animal("GenericAnimal")
a.speak()

d = Dog("Buddy")
d.speak()
```

```python
# Introducing attribute and another class: 
class Animal:
    def __init__(self, name, age):
        self.name = name  # instance variable
        self.age = age
        print(f"{self.name} the animal is created.")

    def speak(self):
        print(f"{self.name} makes a generic animal sound.")

class Dog(Animal):
    def speak(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} says Meow!")

# Let's create some animals
dog1 = Dog("Buddy", 3)
cat1 = Cat("Whiskers", 2)

# Call their methods
dog1.speak()  # Buddy says Woof!
cat1.speak()  # Whiskers says Meow!
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"{self.name} the animal is created.")

    def speak(self):
        print(f"{self.name} makes a generic animal sound.")

    def birthday(self):
        self.age += 1
        print(f"🎉 Happy Birthday, {self.name}! You are now {self.age} years old.")

class Dog(Animal):
    def speak(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} says Meow!")

class Bird(Animal):
    def speak(self):
        print(f"{self.name} says Chirp!")

# Create some animals
dog1 = Dog("Rex", 4)
cat1 = Cat("Mittens", 5)
bird1 = Bird("Tweety", 1)

# Let them speak
dog1.speak()    # Rex says Woof!
cat1.speak()    # Mittens says Meow!
bird1.speak()   # Tweety says Chirp!

# Celebrate birthdays!
dog1.birthday()
cat1.birthday()
bird1.birthday()
```
