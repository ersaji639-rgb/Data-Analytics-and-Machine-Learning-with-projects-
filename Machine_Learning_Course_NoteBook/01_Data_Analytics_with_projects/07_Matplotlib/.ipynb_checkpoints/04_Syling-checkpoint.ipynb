```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0,10)
y = 2**x
```

```python
x
```

```python
y
```

### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label = 'x^2')

ax.plot(x,x**2 + 100, label = 'x^2 +100')
#ax.legend(loc = 'best') # lower right etc
ax.legend(loc = (1.1,0.5)) # custom location  
plt.plot()
plt.show()
```

### Setting colors, line width, line style etc 

```python
# Press shift tab to get the info : 

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label = 'x^2', color = 'green', marker = 'o', linestyle = 'dashed', linewidth = 2, markersize = 12, mfc = 'blue', mec = 'purple',mew = '5.9')

ax.plot(x,x**2 + 100, label = 'x^2 +100', marker = 'v',ms = 12)
#ax.legend(loc = 'best') # lower right etc
ax.legend(loc = (1.1,0.5)) # custom location  
plt.plot()
plt.show() 

 
```
