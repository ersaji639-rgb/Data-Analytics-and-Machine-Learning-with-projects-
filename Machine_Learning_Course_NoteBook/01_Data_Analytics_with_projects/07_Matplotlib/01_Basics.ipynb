```python
import numpy as np
import matplotlib.pyplot as plt
```

### Using function based methods

```python
x = np.arange(0,10)
print(x)
```

```python
y = 2*x 
print(y)
```

### Basic matplotlib commands

```python
plt.plot(x,y)
plt.xlabel('X-axis title')
plt.ylabel('Y-axis labels')
plt.title('This Figure Title')
plt.xlim(2,6)
plt.ylim(2.5,12.5)
plt.savefig('Demo.png')
plt.show() # it will remove the output cell and print the plot
```
