```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10,10)) # this is 1 array 
axes[0,0].plot(a,b)
axes[0,0].set_title('First plot')
axes[0,0].set_xlim(2.5,7.5)
axes[0,0].set_xlabel('x-axis')
axes[0,0].set_ylabel('y-axis')
axes[0,0].set_ylim(2000,4000)
axes[0,0].set_yticks([2000, 3000, 4000])

axes[0,1].plot(y,x)
axes[1,1].plot(b,a)
axes[1,0].plot(x,y)


fig.suptitle('Figure level title')
fig.supxlabel('Figure level x axis')
fig.supylabel('Figure level y axis')

plt.savefig('demo3.png')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)

for i, ax in enumerate(axes.flat):
    ax.plot(x, y)
    ax.set_title(f'Plot {i+1}')
    y = y**2  # Just to change data for demonstration
```

### Use of enumerate 


```python
# with out enumerate 

mylist = ['a', 'b', 'c']
for i in range(len(mylist)):
    print(i, mylist[i])

```

```python
# with enumerare
for i, val in enumerate(['a', 'b', 'c']):
    print(i, val)
```

```python
# loop using for loop
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)

count = 1  # to number the plots

for i in range(2):         # rows
    for j in range(2):     # columns
        ax = axes[i, j]    # access subplot at row i, col j
        ax.plot(x, y)
        ax.set_title(f'Plot {count}')
        y = y**2           # modify y for the next plot
        count += 1
```

```python
# Using enumerare and flatten 
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)

for i, ax in enumerate(axes.flatten()):
    ax.plot(x, y)
    ax.set_title(f'Plot {i+1}')
    y = y**2  # modify y for each subplot
```
