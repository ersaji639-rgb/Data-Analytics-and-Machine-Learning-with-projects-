```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
x
```

```python
y = x**4
y
```

```python
a = np.arange(0,10)
a
```

```python
b = 2*a
b
```

### Create a figure

```python
fig = plt.figure() # create a blank canvas


# Add set of axes to figure 
axes = fig.add_axes([0,0,1,1])  # [coord of left bottom corner, top right corner coord ], note this is the size of canvas

# Plot the axes 
axes.plot(x,y)
plt.show()
```

### Add another set of axes to the figure

```python
fig = plt.figure(figsize = (10,10),dpi= 200 ) # create a blank canvas

# Add set of axes to figure 
axes1 = fig.add_axes([0,0,1,1])  # [coord of left bottom corner, top right corner coord ], note this is the size of canvas like a percentage 
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])  # it is like a size in percentage. 
axes3 = fig.add_axes([1.2,1.2,0.4,0.4])
# Plot the axes 
axes1.plot(x,y)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_xlim(2,8)
axes1.set_ylim(1000,8000)
axes1.set_title('Bigger Figure')

axes2.plot(a,b)
axes2.set_xlabel('a label')
axes2.set_ylabel('b label')
axes2.set_title('Smaller Figure')


axes3.plot(y,x)
plt.savefig('demo2.png', bbox_inches = 'tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
a = np.linspace(2, 4, 50)
b = np.sin(a)

# Create figure
fig = plt.figure(figsize=(8, 6),dpi = 200)

# Main axes
axes1 = fig.add_axes([0.1, 0.1, 0.85, 0.85])
axes1.plot(x, y, color='blue')
axes1.set_xlabel('X label', fontsize=12)
axes1.set_ylabel('Y label', fontsize=12)
axes1.set_title('Main Plot', fontsize=14)

# Set ticks for main plot
axes1.set_xticks(np.arange(0, 11, 2))       # Tick every 2 units
axes1.set_yticks(np.arange(-1, 1.1, 0.5))    # Tick every 0.5 units
axes1.tick_params(axis='both', labelsize=10)  # Set font size for tick labels
axes1.tick_params(direction='in', length=6, width=1)
# Inset axes
axes2 = fig.add_axes([0.6, 0.55, 0.3, 0.3])
axes2.plot(a, b, color='red')
axes2.set_xlabel('a', fontsize=10)
axes2.set_ylabel('b', fontsize=10)
axes2.set_title('Inset Plot', fontsize=10)

# Set ticks for inset plot
axes2.set_xticks(np.linspace(2, 4, 5))     # 5 ticks between 2 and 4
axes2.set_yticks(np.linspace(0.9, -1, 5))  # Example range
axes2.tick_params(axis='both', labelsize=8, direction = 'in')

# Optional: Highlight inset region in main plot
axes1.axvspan(2, 4, color='gray', alpha=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# Use mathtext (not full LaTeX) for compatibility
rcParams.update({
    'font.family': 'serif',
    'font.size': 12,
    'axes.linewidth': 1,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.major.width': 1,
    'ytick.major.width': 1,
    'text.usetex': False  # Important fix here
})

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
a = np.linspace(2, 4, 50)
b = np.sin(a)

# Create figure
fig = plt.figure(figsize=(6, 4.5))

# Main plot
axes1 = fig.add_axes([0.12, 0.12, 0.82, 0.82])
axes1.plot(x, y, color='black', linewidth=1)
axes1.set_xlabel(r'$x$', fontsize=12)
axes1.set_ylabel(r'$\sin(x)$', fontsize=12)
axes1.set_xticks(np.arange(0, 11, 2)) # here also if u manually eep the values in $$ then it becomes latex
axes1.set_yticks(np.arange(-1, 1.1, 0.5))
axes1.tick_params(axis='both', labelsize=10)
axes1.set_title('Main Plot', fontsize=13)
axes1.axvspan(2, 4, color='gray', alpha=0.2, zorder=0)

# Inset plot
axes2 = fig.add_axes([0.55, 0.52, 0.35, 0.35])
axes2.plot(a, b, color='red', linewidth=1)
axes2.set_xlabel(r'$a$', fontsize=10)
axes2.set_ylabel(r'$\sin(a)$', fontsize=10)
axes2.set_xticks(np.linspace(2, 4, 5))
axes2.set_yticks(np.linspace(-1, 1, 5))
axes2.tick_params(axis='both', labelsize=8)

for spine in axes2.spines.values():
    spine.set_linewidth(0.8)

plt.savefig("journal_plot.pdf", bbox_inches='tight', dpi = 600)  # Save as PDF
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Journal-style settings
rcParams.update({
    'font.family': 'serif',
    'font.size': 12,
    'axes.linewidth': 1,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.major.width': 1,
    'ytick.major.width': 1,
    'mathtext.fontset': 'cm',
    'text.usetex': False  # Set to True only if LaTeX is installed
})

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
a = np.linspace(2, 4, 50)
b = np.sin(a)

# Create main plot using subplots
fig, ax = plt.subplots(figsize=(6, 4.5))

# Main plot
ax.plot(x, y, color='black', linewidth=1)
ax.set_xlabel(r'$x$', fontsize=12)
ax.set_ylabel(r'$\sin(x)$', fontsize=12)
ax.set_xticks(np.arange(0, 11, 2))
ax.set_yticks(np.arange(-1, 1.1, 0.5))
ax.tick_params(axis='both', labelsize=10)
ax.set_title('Main Plot', fontsize=13)
ax.axvspan(2, 4, color='gray', alpha=0.2, zorder=0)

# Create inset using inset_axes
ax_inset = inset_axes(ax, width="35%", height="35%", loc='upper right',
                      bbox_to_anchor=(0, 0, 1, 1), bbox_transform=ax.transAxes)

ax_inset.plot(a, b, color='red', linewidth=1)
ax_inset.set_xlabel(r'$a$', fontsize=10)
ax_inset.set_ylabel(r'$\sin(a)$', fontsize=10)
ax_inset.set_xticks(np.linspace(2, 4, 5))
ax_inset.set_yticks(np.linspace(-1, 1, 5))
ax_inset.tick_params(axis='both', labelsize=8)

# Optional: thin border
for spine in ax_inset.spines.values():
    spine.set_linewidth(0.8)

# Save high-res
plt.savefig("journal_plot_subplots.pdf", bbox_inches='tight', dpi=600)
plt.show()
```
