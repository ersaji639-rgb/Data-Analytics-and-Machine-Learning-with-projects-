### Enumurate

```python
index_count = 0
for  let in "Hello world":
    print(f"At index {index_count}, the letter is {let}") 
    index_count += 1 # used to count the loop exectution time
```

```python
for i,l in enumerate('Hello world'):
        print(f"At index {i}, the letter is {l}") 
```

* enumerate is creating bunch of     tuple: note:  elements in a tuple cannot be changed once the tuple is created. Tuples in Python are immutable, meaning their contents are fixed.

```python
enumerate('Hello world')
```

```python
# Cast it in to a list 
list(enumerate('Hello world'))
```

```python
#You can create a new tuple with modified values:

my_tuple = (1, 2, 3)
new_tuple = (10,) + my_tuple[1:]
print(new_tuple)  # Output: (10, 2, 3)
```

```python
t = ([1, 2], 3)
t[0].append(4)  # ✅ This is allowed
print(t)        # Output: ([1, 2, 4], 3)
```

### zip

```python
list1 = [11,12,13,14,15]
list2 = ['a1','b1','c1','d1','e1']
list(zip(list1,list2))  # note bothlist should beof same length 
```

```python
tuple(zip(list1,list2)) 
```

```python
for x,y in zip(list1,list2):
    print(f"For this tuple, first item {x} comes from list 1, second item from list {y} from list 2")
```
