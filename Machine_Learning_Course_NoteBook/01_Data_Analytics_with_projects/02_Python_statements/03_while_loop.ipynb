 Q5 – Reverse Multiples

Using a while loop:

    Print numbers from 100 down to 1

    But:

        If divisible by 4, print "Quad"

        If divisible by 6, print "Hex"

        If divisible by both → print "QuadHex"
        

```python
i = 100

while i >= 1:
    if i % 4 == 0 and i % 6 == 0:
        print("QuadHex")
    elif i % 4 == 0:
        print("Quad")
    elif i % 6 == 0:
        print("Hex")
    else:
        print(i)
    i -= 1
```

 Q6 – Word Length Classifier

Given a sentence, classify each word as:

    "Short" if it's ≤3 letters

    "Medium" if it's 4–6 letters

    "Long" if it's >6 letters

📦 Bonus: Store them in separate lists and print the lists at the end.

```python
st = "Python is fun and sometimes a little challenging"

short_words = []
medium_words = []
long_words = []

words = st.split()

for word in words:
    l = len(word)
    if l <= 3:
        short_words.append(word)
    elif 4 <= l <= 6:
        medium_words.append(word)
    else:
        long_words.append(word)

print("Short words:", short_words)
print("Medium words:", medium_words)
print("Long words:", long_words)
```

### Check wheather a num is palindrome or not

```python

def is_palindrome(value):
    # Convert to string in case it's a number
    value_str = str(value)
    # Remove spaces and convert to lowercase for case-insensitive check
    value_str = value_str.replace(" ", "").lower()
    # Check if the string is the same when reversed
    return value_str == value_str[::-1]

# Examples
print(is_palindrome("Madam"))         # True
print(is_palindrome("racecar"))       # True
print(is_palindrome("Hello"))         # False
print(is_palindrome(121))             # True
print(is_palindrome(12321))           # True
```

```python
st = input('Enter a string to check: ')
st = st.lower()
st_len = len(st)
st_check = '' # Using string (can be addable)

# Build the reversed string manually
for i in range(st_len - 1, -1, -1):  # Loop from end to start
    st_check += st[i]

# Check if original and reversed are the same
if st == st_check:
    print("It's a palindrome!")
else:
    print("Not a palindrome.")
```

```python
# Using append If u use list u can   append:
st = input('Enter a string to check: ')
st = st.lower()
st_len = len(st)

# Use a list to build the reversed string
st_check_list = []

# Append characters in reverse order
for i in range(st_len - 1, -1, -1):
    st_check_list.append(st[i])

# Convert list back to string
st_check = ''.join(st_check_list)

# Check if it's a palindrome
if st == st_check:
    print("It's a palindrome!")
else:
    print("Not a palindrome.")
```

Key differences between isdigit() and isnumeric():

    isdigit(): Checks if all characters are digits (0-9). It does not allow decimal points or negative signs.

    isnumeric(): Checks if all characters are numeric, which includes digits and some special characters like ², ³, etc. It can handle more than just the usual digits.
    

```python
s = "456"
print(s.isdigit())  # Output: True
```

```python
s = "45.6"
print(s.isdigit())  # Output: False (because of the dot)
```

```python
s = "45.6"
print(s.isnumeric())  # Output: False (because of the dot)
```

```python
s1 = "²"  # Superscript two
s2 = "³"  # Superscript three

print(s1.isdigit())   # Output: False
print(s1.isnumeric()) # Output: True

print(s2.isdigit())   # Output: False
print(s2.isnumeric()) # Output: True
```

```python
s1 = "Ⅷ"  # Roman numeral for 8
s2 = "Ⅸ"  # Roman numeral for 9

print(s1.isdigit())   # Output: False
print(s1.isnumeric()) # Output: True

print(s2.isdigit())   # Output: False
print(s2.isnumeric()) # Output: True
```

```python
s1 = "-123"  # A negative number

print(s1.isdigit())   # Output: False (because of the minus signm
print(s1.isnumeric()) # Output: False (because of the minus sign)
```

```python
s1 = "123.45"  # A floating point number

print(s1.isdigit())   # Output: False (because of the dot)
print(s1.isnumeric()) # Output: False (because of the dot)
```

ummary:

    isdigit(): Works only for characters that are digits (0-9). It fails for numbers with signs, decimals, or special numeric characters like superscripts.

    isnumeric(): Works for a broader range of numeric characters, including superscripts, fractions, and some special numeric symbols, but fails for decimals or negative signs.m
