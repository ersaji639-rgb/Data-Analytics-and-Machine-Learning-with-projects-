### 🔢 Use range(start, stop, step)

```python
for i in range(2, 10, 2):  # from 2 to 8 with step of 2
    print(i)
```

```python
for i in range(2, 11, 2):
    print(i)
```

### Looping through chars in a string

```python
for letter in "Saji":
    print(letter)
```

### Optional: Skip or Stop with continue and break

```python
for i in range(5):
    if i == 2:
        continue  # skip 2
    print(i) # note this print statement is comming under for loop. 
```

```python
for i in range(5):
    if i == 3:
        break  # stop when i is 3
    print(i)
```

```python
for i in range(1, 6):  # From 1 to 5
    print(i)
```

```python
for i in range(6):  # From 1 to 5
    print(i)
```

```python
for i in range(1, 6, 1):  # From 1 to 5
    print(i)
```

###  2. Counting Down

```python
for i in range(5, 0, -1):  # From 5 to 1
    print(i)
```

```python
# tuple 
t1 = {1,2,3,4,5}
for n in t1:
    print(n)
```

```python
#tuple unpacking 
l5  = [{2,4},{6,8},{10,12}] # list containing tuples
```

```python
for x in l5:
    print(x)
```

```python
# tuple packing 
for x1,x2 in l5:
    print(x1+x2)
```
