* grab the first letter of the string  with eevn length : split--> check length --> grab the first letter.

* Filter out all the vowels from the string original string  and keep only consonants.

* create a list of numbers from 1 - 20 where each number is replaced with "even" and "odd" for od


```python
text = "Python is a powerful and elegant programming language"
words = text.split()

# Get the first letter of words with even length
first_letters = [word[0] for word in words if len(word) % 2 == 0]
print(first_letters)
```

```python
text = "Beautiful day in the neighborhood"
vowels = "aeiouAEIOU"

# Keep only consonants
only_consonants = ''.join([char for char in text if char not in vowels])
print(only_consonants)
```

```python
even_odd_list = ["even" if num % 2 == 0 else "odd" for num in range(1, 21)]
print(even_odd_list)
```

 What is List Comprehension?

It's a shorthand for:

new_list = [expression for item in iterable if condition]

It replaces this longer format:

new_list = []
for item in iterable:
    if condition:
        new_list.append(expression)


```python
even_odd_list = []
for num in range(1, 21):
    if num % 2 == 0:
        even_odd_list.append("even")
    else:
        even_odd_list.append("odd")
print(even_odd_list)
```

```python
even_odd_list = ["even" if num % 2 == 0 else "odd" for num in range(1, 21)]
print(even_odd_list)
```
