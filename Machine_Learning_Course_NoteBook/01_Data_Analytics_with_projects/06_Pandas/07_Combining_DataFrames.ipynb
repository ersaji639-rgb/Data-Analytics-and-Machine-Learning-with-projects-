### Merge

```python
reg = pd.DataFrame({'reg_id':[1,2,3,4],
                   'name':['Andrew','Bob','Charlie','David']})
log = pd.DataFrame({'log_id':[1,2,3,4],
                   'name':['Xavier','Andrew','Yolanda','Bob']})
```

```python
reg
```

```python
log
```

```python
help(pd.merge) # Read      this on
```

### Inner

```python
pd.merge(reg,log,how = 'inner', on = 'name') # Common 
```

```python
pd.merge(reg,log,how = 'inner') # Common 
```

### Left

```python
pd.merge(reg,log,how = 'left', on = 'name')
```

### Right 

```python
pd.merge(reg,log,how = 'right', on = 'name') 
```

###  outer 

```python
pd.merge(reg,log,how = 'outer', on = 'name') # all 
```

### Dealing with different key column names 

```python
reg = pd.DataFrame({'reg_id':[1,2,3,4],
                   'reg_name':['Andrew','Bob','Charlie','David']})
log = pd.DataFrame({'log_id':[1,2,3,4],
                   'name':['Xavier','Andrew','Yolanda','Bob']})
```

```python
reg
```

```python
log
```

```python
pd.merge(reg, log, left_on = 'reg_name', right_on = 'name')
```

```python
pd.merge(reg, log, left_on = 'reg_name', right_on = 'name').drop('reg_name', axis = 1)
```
