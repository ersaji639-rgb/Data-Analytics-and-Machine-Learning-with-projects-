### Scatter plot

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['D']*50)
```

```python
# the above warning is due to -ve number 

```

```python
# option 1 : 1: Use the absolute value
df1.plot.scatter(x='A', y='B', s=(df1['D'].abs() * 50))
```

```python
# option 2 :  Clip negative or zero values (ensure positive sizes) 
df1.plot.scatter(x='A', y='B', s=(df1['D'].clip(lower=0) * 50))
```

```python
# Option 3: Handle NaNs if needed 
df1_clean = df1.dropna(subset=['A', 'B', 'D'])
df1_clean.plot.scatter(x='A', y='B', s=(df1_clean['D'].abs() * 50), cmap='viridis')
```
