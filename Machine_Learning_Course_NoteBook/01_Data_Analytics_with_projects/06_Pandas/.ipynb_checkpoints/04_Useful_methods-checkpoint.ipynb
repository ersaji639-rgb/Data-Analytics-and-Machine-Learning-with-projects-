###  Apply method 

```python
def sq(num):
    return np.round(num**2,2)
```

```python
sq(3.4)
```

```python
df['tip'].apply(sq)  # we squared all the numbers in the tip 
```

```python
# To find len of string 
def length(s):
    return len(s) 
```

```python
df["Payer Name"].apply(length)
```

```python
df.head()
```

```python
 #  i want to mask some portion of ctredit card number 
def last_four(num):
    return "*"*12 + str(num)[-4:] # 12 astrists +(concatanate)  last four digit 
```

```python
last_four(45678909764345)
```

```python
df["CC Number"].apply(last_four)
```

```python
df["masked_CC"] = df["CC Number"].apply(last_four)
```

```python
df
```

### Using lambda

```python
 df["CC Number"].apply( lambda num: "*"*12 + str(num)[-4:])
```

```python
# Rate the customer based on bill > 30 : rate them base on some star 

def ratings(price):
    if price < 10:
        return "*"
    elif price >= 10 and price < 30:
        return "**" 
    else:
        return "***"
```

```python
ratings(34)
```

```python
df["total_bill"].apply(ratings)
```

```python
df["customer_kind"] = df["total_bill"].apply(ratings)
```

```python
df
```

```python
df[df["customer_kind"] == "***"]
```

```python
# create a new column gives final bill 
# Total bill + 5% GST 
def final_bill(num):
    return num + 0.05*num
    
```

```python
final_bill(10)
```

```python
df.head()
```

```python
df["total_bill"].apply(final_bill)

#df["final_bill"] = df["total_bill"].apply(lambda x: np.round(1.05*x, 2))
```

```python
df["final_bill"] = df["total_bill"].apply(final_bill)
```

```python
df
```

```python
# df["is_generous" ] being true if tip is > 20 % of total bill else false 
df["is_generous"] = df["tip"] > 0.2 * df["total_bill"]
```
