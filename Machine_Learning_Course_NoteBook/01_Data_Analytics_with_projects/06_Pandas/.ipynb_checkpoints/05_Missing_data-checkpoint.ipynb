#### Checking and Selecting Null Values   

```python
df = pd.read_csv('movie_scores.csv')
```

```python
df
```

```python
df.isnull() # Work in the context of data frame. 
```

```python
df.info()
```

```python
df.isna()
```

```python
df.isna().sum() # To see how many missing values are there. in each column. U can use df.info() also 
```

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
df['first_name']
```

```python
type(df['first_name'])
```

```python
df['first_name'].notnull()
```

```python
df['first_name'].notnull().sum()
```

```python
# Now u can use above as a conditional filtering for entitre data frame 
df[df['first_name'].notnull()] # get all the values for which this first name can be null. 
```

```python
# Conditional filtering based on two columns 
df['pre_movie_score'].isnull()
```

```python
df['sex'].notnull()
```

```python
# I want all thevalues in which the pre movie score is null andsex is not null 
df[df['pre_movie_score'].isnull() & df['sex'].notnull()]
```

```python
df
```

#### Drop Data 


```python
df.dropna() # it drops all the row wihich have NAN value
```

```python
df.dropna(thresh = 1 ) # that is there should be atleast one non null value in that row
```

```python
df.dropna(thresh = 4 )
```

```python
df.dropna(thresh = 5 ) # I need 5 non null values
```

```python

 #I want drom column with missing values (default axis = 0 ie: row wise)  
df.dropna(axis = 1)
```

```python
df.dropna(axis = 1,thresh = 1)
```

```python
df.dropna(axis = 1,thresh = 4) # i want atleast 4 non-null values in the column. 
```

#### Filling with Interpolation

```python
# Dictionary 
rail_tickt = {'first':100,
              'buisness':np.nan,
              'buisness1':np.nan,
              'second':50,
              'economy':30}
```

```python
ser = pd.Series(rail_tickt)
```

```python
ser
```

```python
ser.interpolate() # This series should be ascending or descending for this interpolate to work 
```

### Groupby Operation

It is an operation allows us to examine data on a "per category" basis
categorical column can be numbers as well 
Eg:
Category    Data value 
A          10
A          5
B          4                
B          6

Now pandas segregate this: 
A          10
A          5

B          4                
B          6

we can perform operation separately
like .mean()
.count() # counts num of rows in that category
