```python
df[df["total_bill"] > 25] # it gives all the instances where the total bill is atleast 25
```

```python
df[df["sex"] == "Male"] # To filter all male customer in our data frame
```

### Multiple conditions

* AND ----> &
* OR ---> | # pipe operator
* NOT ----> ~ 

```python
df[(df["tip"]>5) & (df["sex"] == "Female")]  # that is female customer who paid the tip > 7 
```

```python
# Q1 I want to get all the records where tip is < 2 and total bill is above 10

df[(df["tip"] < 2) & (df["total_bill"] > 10)]
```

```python
# I want all customers list who visited resturant on weekend 
# Filter for weekend days
#weekend_df = df[df['day'].isin(['Sat', 'Sun'])]

# Optionally, show just the customer names
#customers = weekend_df['Payer Name'].tolist()

#print(customers)
df[df['day'].isin(['Sat', 'Sun'])]
```

```python
df[(df["day"] == "Sat" ) | (df["day"] == "Sun")]
```

```python
# to verify 
df1 = df[(df["day"] == "Sat" ) | (df["day"] == "Sun")] 
df1["day"].value_counts()
```

```python
# or the above can be done by 
# Filter for weekend days
weekend_df = df[df['day'].isin(['Sat', 'Sun'])]

# Optionally, show just the customer names
customers = weekend_df['Payer Name'].tolist()

print(customers)
```

```python
#I want all the smokers  who visted on thursday and bill is is less than 15 
df[(df["smoker"] == "Yes") & (df["total_bill"] < 15) & (df["day"].isin(["Thur"]))]
```

```python
#I want all the entries where customers are either not female or give a tip higher that 6. 
#All rows where:

   # the customer is male, or

   # the tip they gave is more than 6, even if they are female.

df[(df["sex"] != "Female") | (df["tip"] > 6)]
```

```python
# to verify 
df2 = df[(df["sex"] != "Female") | (df["tip"] > 6)]

df2["sex"].value_counts()
```

```python
 df[~(df["sex"] == "Female") | (df["tip"] > 6)]
```
