# doing it by list 
To do something similar with a Python list:

You cannot assign a single value directly to a slice like NumPy (broadcasting), but you can do it using slicing and repetition:m

```python
lst = list(range(20, 71, 3))
lst[0:5] = [100] * 5  # Replace first 5 elements with 100
print(lst)
```

Summary:

    NumPy allows broadcasting: arr[0:5] = 100

    Python lists need explicit repetition: lst[0:5] = [100] * 5
