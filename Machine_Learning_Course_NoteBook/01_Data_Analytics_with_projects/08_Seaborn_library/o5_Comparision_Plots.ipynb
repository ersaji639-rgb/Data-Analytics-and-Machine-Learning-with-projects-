```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv') 
df.head()
```

```python
df.columns
```

### Jointplot

```python
sns.jointplot(x = 'writing score',y = 'reading score', data = df ) # scatter and histogram together 
```

```python
sns.jointplot(x = 'writing score',y = 'reading score', data = df, kind = 'hex' )
```

```python
sns.jointplot(x = 'writing score',y = 'reading score', data = df, kind = 'kde' )
```

```python
sns.jointplot(x = 'math score',y = 'reading score', data = df, kind = 'kde' )
```

### pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'gender')
```

```python
sns.pairplot(df, hue = 'gender', diag_kind = 'hist', corner = False)
```
