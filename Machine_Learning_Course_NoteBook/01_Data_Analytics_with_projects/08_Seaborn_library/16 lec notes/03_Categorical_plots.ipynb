```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

###### Countplot

```python

plt.figure(figsize=(12,6))
sns.countplot(data=df,x='division')
```

```python
sns.countplot(data=df,x='division',hue='level of education')
plt.xticks(rotation=90);
```

###### Barplot

```python
plt.figure(figsize=(12,6))
sns.barplot(x='level of education',y='salary',data=df,estimator='mean',hue='division')
```
