```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

###### Scatterplot

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y='salary',x='sales',data=df)
plt.plot()
```

###### Seaborn parameters

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y='salary',x='sales',data=df,hue = 'division')
plt.legend(loc=(1.2,0.5))
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y='salary',x='sales',data=df,size = 'training level',palette='Set1',hue='work experience')

#use s (instead of size) for consistent marker size. It accepts integer as parameter
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y='salary',x='sales',data=df,size = 'training level',color='red',alpha=0.2)
```

```python
#style

plt.figure(figsize=(12,6))
sns.scatterplot(y='salary',x='sales',data=df,style='level of education', size = 'training level', hue = 'division')
plt.legend(loc=(1.01,0.1))
plt.savefig('example_scatter.jpg',bbox_inches='tight')
```
