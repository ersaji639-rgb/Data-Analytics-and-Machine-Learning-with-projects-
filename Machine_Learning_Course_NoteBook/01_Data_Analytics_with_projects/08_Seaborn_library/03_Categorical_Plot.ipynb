```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

### countplot()

```python
plt.figure(figsize=(12,6))
sns.countplot(data= df, x = 'division')
```

```python
sns.countplot(data= df, x = 'division')
plt.xticks(rotation = 45);
```

```python
sns.countplot(data= df, x = 'division', hue = 'level of education')
plt.xticks(rotation = 45);
```

###  barplot

```python
sns.barplot(x = 'level of education', y = 'salary', data = df)
plt.xticks(rotation = 45)
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df) # by default this estimator gives mean 
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.std) # For std of salary on each section 
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.var) # For var
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.sum)
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = sum)
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = 'mean')
plt.xticks(rotation = 45); 
```

```python
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = 'mean',hue = 'division')
plt.xticks(rotation = 45); 
```

### Categorical plot - 2
* To compare the distribution across category
* Boxplot -- displays the istribution of continous var. Separates in to 4 quartiles. 
* Violinplot
* Swarmplot
## Boxplot: 
	* Meaning
* Box (rectangle)	From Q1 (25th percentile) to Q3 (75th percentile) — the IQR
* Line inside box	Median (50th percentile)
* Whiskers	Typically extend to data within 1.5×IQR from Q1 and Q3
* Dots outside whiskers	Outliers

* Why it's used in categorical plots:

 * It allows you to compare distributions between different categories.

 *   Shows skewness, spread, and outliers in each group.

 *   It's compact and more informative than bar plots with just means.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title("Total Bill Distribution by Day")
plt.show()
```
