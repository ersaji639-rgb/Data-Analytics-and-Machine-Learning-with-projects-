###  Seaborn parameters

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, hue = 'division')
plt.legend(loc = (1.2,0.5))
#plt.plot()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, hue = 'training level')
#plt.legend(loc = (1.2,0.5))
#plt.plot()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, hue = 'training level',palette = 'Set1')
#plt.legend(loc = (1.2,0.5))
#plt.plot()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, hue = 'training level',palette = 'coolwarm') # See documentation for palette 
#plt.legend(loc = (1.2,0.5))
#plt.plot()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, size = 'training level',palette = 'Set1', hue = 'work experience')
plt.legend(loc = (1.2,0.5))
#plt.plot()
#use size is for marker sizedit should be aninteger
```

```python
df.head()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, size = 'training level',palette = 'Purples', hue = 'work experience', alpha = 0.6)
# alpha for density 
plt.legend(loc = (1.2,0.5))
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, size = 'training level',color = 'red', alpha = 0.2) 
plt.legend(loc = (1.2,0.5))
```

### Style

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, style = 'level of education')
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, style = 'level of education',hue = 'training level')
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(y = 'salary', x= 'sales', data = df, style = 'level of education',size = 'training level', hue = 'division')
plt.legend(loc =( 1.05, 0.5))
plt.savefig('scatter.jpg',bbox_inches = 'tight')
```

```python
# Histogram 
# Bin size should be same 
# count the number of ticks in a bin and increase the height. 
# This is just extended version of rug plot (in rug plot the ticks are of same height, hence it doesnt show how many number of person in that particu;ar salary ragne for eg) 
# sea born allows as to plot the KDE line : its shows the continous distribition which follows the trend of Histogram. 
# KDE estimating the prob density fun of a random variable. 
# kernel could be gaussion, normal
# eevery tick point has gaussion curve centered at that point and add them up at every tick point : it will give a curve from KDE plot. 
# We can always change the kernel : not necessarly to be Gaussian. 
```
