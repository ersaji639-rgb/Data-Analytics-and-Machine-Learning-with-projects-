```python
def spy_game(nums):
    code = [0, 0, 7]  # target sequence
    for num in nums:
        if num == code[0]:
            code.pop(0)  # remove matched number from the sequence
        if not code:
            return True  # sequence matched completely
    return False  # sequence not matched

# Example usage
print(spy_game([1, 2, 4, 0, 0, 7, 5]))  # True
print(spy_game([1, 0, 2, 4, 0, 5, 7]))  # True
print(spy_game([1, 7, 2, 0, 4, 0, 5]))  # False
```
