### any

```python
any([False, False, True])   # returns True
any([0, 0, 1])              # returns True (because 1 is treated as True)
any([False, False, False])  # returns False
```

```python
s = "abc123"
any(char.isdigit() for char in s)
#any([False, False, False, True, True, True])  # ths will be the op
```

```python
# Like this behind the scenes:
[char.isdigit() for char in s]
# returns: [False, False, False, True, True, True]

# any() checks that list:
any([False, False, False, True, True, True])  # returns True
```

```python
# Exersice : . Basic Boolean List
print(any([False, False, True]))  # True (because at least one is True)
print(any([False, False, False])) # False (no True values)
```

```python
#  2. With numbers (remember: 0 is False, non-zero is True) 
print(any([0, 0, 1]))     # True (because 1 is considered True)
print(any([0, 0, 0]))     # False
```

```python
# ✅ 3. With strings (non-empty is True, empty string is False) 
print(any(["", "", "hello"]))   # True
print(any(["", "", ""]))        # False
```

```python
# ✅ 4. With a generator and condition — check for even numbers 
nums = [1, 3, 5, 8]

# Check if there's any even number
print(any(n % 2 == 0 for n in nums))  # True (because 8 is even)
```

```python
# ✅ 5. Check if a string has any digit 
s = "abc1"
print(any(char.isdigit() for char in s))  # True (because '1' is a digit)

s2 = "hello"
print(any(char.isdigit() for char in s2))  # False (no digits)
```

 1. lambda (anonymous function)

A lambda is just a tiny function you define without a name.

```python
# Normal function
def square(x):
    return x*x

print(square(5))  # 25

# Same thing with lambda
square = lambda x: x*x
print(square(5))  # 25
```

 2. map() – Apply a function to each item in a list

```python
nums = [1, 2, 3, 4]

# Double each number
doubled = list(map(lambda x: x*2, nums))
print(doubled)  # [2, 4, 6, 8]
```

🔹 3. filter() – Keep items that match a condition

```python
nums = [1, 2, 3, 4, 5, 6]

# Keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # [2, 4, 6]
```
