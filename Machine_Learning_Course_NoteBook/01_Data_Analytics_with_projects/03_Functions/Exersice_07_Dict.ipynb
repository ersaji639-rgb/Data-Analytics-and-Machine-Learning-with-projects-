🧑‍🏫 What is a Dictionary in Python?

A dictionary in Python is a collection of key-value pairs. Each key in a dictionary is unique, and it is used to access the corresponding value.

```python
person = {
    "name": "Saji",
    "age": 25,
    "city": "New York"
}
```

```python
# Accessing Values  in dict 
print(person["name"])  # Output: Saji
print(person["age"])   # Output: 25
print(person["city"])  # Output: New York
```

```python
# add 
person["job"] = "Developer"
print(person)
# Output: {'name': 'Saji', 'age': 25, 'city': 'New York', 'job': 'Developer'}
```

```python
# Update 
person["age"] = 26
print(person)
# Output: {'name': 'Saji', 'age': 26, 'city': 'New York', 'job': 'Developer'}
```

```python
# Check
print("age" in person)   # Output: True
print("address" in person)  # Output: False
```

```python
# Remove 
del person["job"]
print(person)
# Output: {'name': 'Saji', 'age': 26, 'city': 'New York'}
```

```python
# Looping through ict
for key in person:
    print(key)
```

```python
# Looping through ict
for value in person:
    print(value)
```

```python
# Looping through values
for value in person.values():
    print(value)
```

```python
quiz = {
    "What is the capital of France?": {
        "options": ["A) Berlin", "B) Madrid", "C) Paris", "D) Rome"],
        "answer": "C"
    }
}
```

```python
quiz.keys()
```

```python
quiz.values()
```

```python
quiz.items()
```

```python
print(quiz["What is the capital of France?"]["options"])
print(quiz["What is the capital of France?"]["answer"])
```

```python
for question, content in quiz.items():
    print(question)  # prints the question
    for option in content["options"]:
        print(option)  # prints each option
```

```python
content["options"] # this is a list 
```

```python
score = 0

for question, content in quiz.items():
    print(question)
    for option in content["options"]:
        print(option)
    
    user_ans = input("Your answer (A/B/C/D): ").strip().upper()

    if user_ans == content["answer"]:
        print("Correct ✅\n")
        score += 1
    else:
        print(f"Wrong ❌, Correct answer is {content['answer']}\n")

print(f"Your final score: {score}/{len(quiz)}")
```

```python
# Full Example : 
quiz = {
    "What is the capital of France?": {
        "options": ["A) Berlin", "B) Madrid", "C) Paris", "D) Rome"],
        "answer": "C"
    },
    "Which number is even?": {
        "options": ["A) 1", "B) 3", "C) 5", "D) 4"],
        "answer": "D"
    }
}

score = 0

for question, content in quiz.items():
    print(question)
    for option in content["options"]:
        print(option)
    
    user_ans = input("Your answer (A/B/C/D): ").strip().upper()

    if user_ans == content["answer"]:
        print("Correct ✅\n")
        score += 1
    else:
        print(f"Wrong ❌, Correct answer is {content['answer']}\n")

print(f"Your final score: {score}/{len(quiz)}")
```

🔹 strip() in Python

strip() is a string method that removes any leading (start) and trailing (end) whitespace characters (like spaces, tabs, or newline \n) from a string.

```python
s = "  hello Saji  "
print(s.strip())  # Output: "hello" # Removes white spaces at start and end. 
```

```python
# Nested Dictionary: 
students = {
    "s1": {"name": "Saji", "age": 21},
    "s2": {"name": "Alex", "age": 22}
}
print(students["s2"]["name"])  # Output: Alex
```

```python
students.items()
```

```python
# Frequency of char count 
word = "hello"
char_count = {}

for char in word:
    if char in char_count:
        char_count[char] += 1  # increment count if char already exists
    else:
        char_count[char] = 1   # add char with count 1 if new

print(char_count)
```
