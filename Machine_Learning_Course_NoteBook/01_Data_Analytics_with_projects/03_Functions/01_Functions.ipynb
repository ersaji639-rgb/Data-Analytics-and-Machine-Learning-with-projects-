```python
def reverse_words(sentence):
    split_str = sentence.split()
    rev_list = []
    for i in range(len(split_str) - 1, -1, -1):
        rev_list.append(split_str[i])
    return ' '.join(rev_list)

# Example usage:
st = "I am Good"
print(reverse_words(st))  # Output: "Good am I"
```

```python
# Given a list return True if the array contains 3 next to 3 somewhere
#[1,3,3,4]--> true, [1,2,3,4,3] ---> false
```

```python
def has_33(nums):
    for i in range(len(nums) - 1):# it will loop unitil end -1 
        if nums[i] == 3 and nums[i + 1] == 3:
            return True
    return False

# Test cases
print(has_33([1, 3, 3, 4]))       # True
print(has_33([1, 2, 3, 4, 3]))    # False
```

```python
def spy_game(nums):
    code = [0, 0, 7]
    code_index = 0
    
    for num in nums:
        if num == code[code_index]:
            code_index += 1
        if code_index == len(code):
            return True
    return False
```

```python
def spy_game(nums):
    code = [0, 0, 7]
    for num in nums:
        if num == code[0]:
            code.pop(0)
        if not code: # In Python, if not code: checks whether the list code is empty.
            return True
    return False
```

```python
print(spy_game([1,2,3,0,0,7,4]))       # True
print(spy_game([1,0,5,0,6,5,7]))       # True
print(spy_game([7,1,3,0,0,4,5]))       # False
```

```python
def primes(num):
    if num < 2:
        return f"There are 0 prime numbers between 0 and {num}"
    
    primes = [2]  # 2 is the only even prime number
    x = 3  # start checking from 3 with step size of 2

    while x <= num:
        for n in range(3, x, 2):  # test all odd factors up to x-1
            if x % n == 0:
                break
        else:  # this else belongs to the for-loop, not the if!
            primes.append(x)
        x += 2  # move to the next odd number

    print(primes)
    return f"There are {len(primes)} prime numbers between 0 and {num}"
```

```python
def primes(num):
    if num < 2:
        return f"There are 0 prime numbers between 0 and {num}"
    
    primes = [2]  # 2 is the only even prime number
    x = 3  # start checking from 3, step by 2 (only odd numbers)

    while x <= num:
        for n in range(3, int(x**0.5) + 1, 2):  # check up to sqrt(x)
            if x % n == 0:
                break
        else:
            primes.append(x)
        x += 2  # move to next odd number

    print(primes)  # for debugging
    return f"There are {len(primes)} prime numbers between 0 and {num}"

```
