* syntax
###  [expression for item in iterable]


🎯 Step 1: From normal loop to list comprehension
🔹 Using loop:

```python
nums = []
for i in range(5):
    nums.append(i)
print(nums)
```

```python
# Using list comprehension 
nums = [i for i in range(5)]
print(nums)
```

```python
# Get even number from 0 to 9 
# list comprhension (Use of loops)  
evens = [i for i in range(10) if i % 2 == 0]
print(evens)
```

```python
# Square of numbers 
squares = [i**2 for i in range(5)]
print(squares)
```

```python
# Step 4: Add if...else in expression
# 🔹 Label even and odd 
labels = ['even' if i % 2 == 0 else 'odd' for i in range(5)]
print(labels)
```

```python
#  From another list
# Uppercase all names
```

```python
 names = ['saji', 'alex', 'john']
upper_names = [name.upper() for name in names]
print(upper_names)
 
```

```python
# Get all numbers divisible by 3 
div_by_3 = [i for i in range(1, 21) if i % 3 == 0]
print(div_by_3)
```

```python
#  Filter strings that contain a digit 
st = "12Saji"
st.isdigit()
```

```python
words = ['apple', 'ban4na', 'pear3', 'grape']
words_with_digit = [word for word in words if any(ch.isdigit() for ch in word)]
print(words_with_digit)
```

 5. List comprehension with nested loops
🔹 Multiply all combinations of (1,2) with (10,20)

```python
products = [i * j for i in [1, 2] for j in [10, 20]]
print(products)
```

```python
# Flattened a nested list 
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(flat)
```
