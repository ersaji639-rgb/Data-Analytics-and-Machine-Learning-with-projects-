### **SSD in K-Means Algorithm**

**SSD** stands for **Sum of Squared Distances** (also called **WCSS – Within-Cluster Sum of Squares**).  
It measures how tightly the data points are grouped inside each cluster.

---

### **Formula**

\[
\text{SSD} = \sum_{k=1}^{K} \sum_{x_i \in C_k} ||x_i - \mu_k||^2
\]

Where:
- \( C_k \) = cluster k  
- \( x_i \) = data point  
- \( \mu_k \) = centroid of cluster k  
- \( ||x_i - \mu_k||^2 \) = squared Euclidean distance

---

### **Why SSD is Important?**

- K-Means tries to **minimize SSD**
- Lower SSD → better compact clusters
- Higher SSD → worse spread-out clusters

---

### **Elbow Method**

SSD is plotted against the number of clusters \(K\):

\[
K \text{ vs SSD}
\]

The “elbow point” indicates the optimal number of clusters.

---

### **Simple Example**

If a cluster contains points `[1, 2, 3]` and its centroid is `2`:

\[
SSD = (1-2)^2 + (2-2)^2 + (3-2)^2 = 1 + 0 + 1 = 2
\]


```python
model.inertia_  # SSD value # K_means is trying to minimize this ssd. 
```

###  Choosing k - Value:

```python
ssd = []
for k in range(2,10):
    model = KMeans(n_clusters= k)
    model.fit(sc_X)

    ssd.append(model.inertia_) 
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd, 'o--',mfc = 'red')
plt.xlabel('K Value')
plt.ylabel('SUM OF SQUARED DISTANCE')
```

```python
pd.Series(ssd).diff().plot(kind = 'bar')
```
